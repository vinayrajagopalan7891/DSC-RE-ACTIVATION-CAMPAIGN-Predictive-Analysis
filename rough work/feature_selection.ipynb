```python
def logistic_regression(column, X_train, X_TEST, Y_train, Y_TEST):
    logreg = LogisticRegression(random_state=123)
    logreg.fit(X_train, X_TEST)
    # predict probabilities
    pred_train = logreg.predict(X_train)
    pred_test = logreg.predict(Y_train)

    # evaluate predictions
    mlp_acc_train = accuracy_score(X_TEST, pred_train)
    mlp_acc_test = accuracy_score(Y_TEST, pred_test)

    mlp_roc_train = roc_auc_score(X_TEST, logreg.predict_proba(X_train)[:,1])
    mlp_roc_test = roc_auc_score(Y_TEST,  logreg.predict_proba(Y_train)[:,1])

    eval_dict = {"Acc_train": mlp_acc_train, "Acc_test": mlp_acc_test, "roc_train":  mlp_roc_train,"roc_test": mlp_roc_test}
    return eval_dict
    
```
